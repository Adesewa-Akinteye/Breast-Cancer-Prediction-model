```python
# To enable plotting graphs in Jupyter notebook
%matplotlib inline 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv(r"C:\Users\hp\Desktop\full stack\Supervised ML resources-20221013T022639Z-001 - Copy\Supervised ML resources\wisc_bc_data.csv")
```

```python
Data.shape
```

```python
Data.dtypes
```

```python
Data['diagnosis'] = Data.diagnosis.astype('category')
```

```python
Data.describe().transpose()
```

```python
Data.groupby(["diagnosis"]).count()
```

drop the first column from the data frame. This is Id column which is not used in modeling

```python
Data = Data.drop(labels = "id", axis = 1)
```

```python
Data.shape
```

```python
X = Data.drop(labels= "diagnosis" , axis = 1)
y = Data["diagnosis"]
X.head()
```

```python
y.head()
```

```python
XScaled  = X.apply(zscore) 

XScaled.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(XScaled, y, test_size=0.30, random_state=1)
```

choosing our k value

```python
scores =[]
for k in range(1,30):
    NNH = KNeighborsClassifier(n_neighbors = k, weights = 'distance' )
    NNH.fit(X_train, y_train)
    scores.append(NNH.score(X_test, y_test))
```

```python
plt.plot(range(1,30),scores)
```

From the above graph, we will get better accuracy from k value between 10 and 15

```python
NNH = KNeighborsClassifier(n_neighbors= 12, weights = 'distance' )
```

```python
NNH.fit(X_train, y_train)
```

```python
predicted_labels = NNH.predict(X_test)
NNH.score(X_test, y_test)
```

```python
from sklearn import metrics

print("Confusion Matrix")
cm=metrics.confusion_matrix(y_test, predicted_labels, labels=["M", "B"])

df_cm = pd.DataFrame(cm, index = [i for i in ["M","B"]],
                  columns = [i for i in ["Predict M","Predict B"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True)
```
